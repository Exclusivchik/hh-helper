```python
import re
import pandas as pd

# ----------------------------
# 1. Таксономия компетенций
# ----------------------------

COMPETENCY_TAXONOMY = {
    "programming": {
        "python": [
            "python",
        ],
        "sql": [
            "sql",
            "postgresql",
            "mysql",
            "clickhouse",
            "bigquery",
        ],
        "scala": [
            "scala",
        ],
        "java": [
            "java",
        ],
        "c++": [
            "c++",
            "cpp",
        ],
        "bash": [
            "bash",
            "shell scripting",
            "linux shell",
        ],
    },

    "ml_dl_core": {
        "machine_learning": [
            "machine learning",
            "машинное обучение",
        ],
        "deep_learning": [
            "deep learning",
            "deep neural networks",
            "neural networks",
            "нейронные сети",
            "глубокое обучение",
        ],
        "feature_engineering": [
            "feature engineering",
            "feature selection",
            "feature extraction",
        ],
        "model_training": [
            "model training",
            "training models",
            "train models",
            "обучение моделей",
        ],
        "model_evaluation": [
            "model evaluation",
            "evaluate models",
            "cross-validation",
            "offline evaluation",
            "валидация моделей",
        ],
        "hyperparameter_tuning": [
            "hyperparameter tuning",
            "parameter tuning",
            "grid search",
            "bayesian optimization",
            "optuna",
        ],
    },

    "llm_nlp": {
        "nlp": [
            "natural language processing",
            "обработка естественного языка",
            "text processing",
            "text mining",
            "nlp",
        ],
        "llm": [
            "large language model",
            "large language models",
            "foundation models",
            "language models",
            "llm",
            "gpt",
            "bert",
            "transformers",
        ],
        "rag": [
            "retrieval augmented generation",
            "retrieval-augmented generation",
            "rag",
        ],
        "prompt_engineering": [
            "prompt engineering",
            "prompt design",
        ],
        "fine_tuning": [
            "fine-tuning",
            "finetuning",
            "instruction tuning",
            "lora",
            "qlora",
        ],
        "embeddings": [
            "embeddings",
            "sentence embeddings",
            "vector search",
            "semantic search",
        ],
        "information_retrieval": [
            "information retrieval",
            "search relevance",
            "reranking",
            "ranking models",
            "retrieval systems",
        ],
        "text_generation": [
            "text generation",
            "summarization",
            "question answering",
            "chatbots",
            "conversational ai",
        ],
    },

    "cv_audio_multimodal": {
        "computer_vision": [
            "computer vision",
            "image processing",
            "object detection",
            "image segmentation",
            "segmentation models",
            "visual recognition",
            "компьютерное зрение",
        ],
        "ocr": [
            "optical character recognition",
            "ocr",
        ],
        "video_ml": [
            "video analytics",
            "video processing",
            "video understanding",
        ],
        "speech": [
            "speech recognition",
            "automatic speech recognition",
            "asr",
            "text to speech",
            "tts",
            "audio processing",
        ],
        "multimodal": [
            "multimodal",
            "vision-language",
            "vision language",
            "vlm",
        ],
    },

    "ml_domains": {
        "recommender_systems": [
            "recommender systems",
            "recommendation systems",
            "recommendation engine",
            "personalized ranking",
            "рекомендательные системы",
        ],
        "time_series": [
            "time series",
            "forecasting",
            "demand forecasting",
            "временные ряды",
            "прогнозирование",
        ],
        "anomaly_detection": [
            "anomaly detection",
            "outlier detection",
            "fraud detection",
        ],
        "causal_inference": [
            "causal inference",
            "uplift modeling",
            "uplift modelling",
        ],
        "experimentation": [
            "a/b testing",
            "ab testing",
            "online experiments",
            "experimentation",
        ],
        "personalization": [
            "personalization",
            "personalisation",
        ],
    },

    "frameworks_libraries": {
        "pytorch": [
            "pytorch",
            "torch",
        ],
        "tensorflow": [
            "tensorflow",
        ],
        "keras": [
            "keras",
        ],
        "scikit_learn": [
            "scikit-learn",
            "sklearn",
        ],
        "xgboost": [
            "xgboost",
        ],
        "lightgbm": [
            "lightgbm",
            "light gbm",
        ],
        "catboost": [
            "catboost",
        ],
        "huggingface": [
            "huggingface",
            "hugging face",
            "transformers library",
        ],
        "opencv": [
            "opencv",
        ],
        "pandas": [
            "pandas",
        ],
        "numpy": [
            "numpy",
        ],
    },

    "data_engineering": {
        "spark": [
            "apache spark",
            "pyspark",
            "spark",
        ],
        "airflow": [
            "apache airflow",
            "airflow",
        ],
        "etl_elt": [
            "etl",
            "elt",
            "data pipelines",
            "pipeline development",
            "data pipeline",
        ],
        "data_warehousing": [
            "data warehouse",
            "data warehousing",
            "snowflake",
            "redshift",
            "bigquery",
        ],
        "streaming": [
            "kafka",
            "apache flink",
            "stream processing",
            "real-time data",
            "streaming data",
        ],
        "feature_store": [
            "feature store",
            "feast",
        ],
    },

    "mlops_production": {
        "mlops": [
            "mlops",
            "ml ops",
        ],
        "model_deployment": [
            "model deployment",
            "deploy models",
            "productionize models",
            "production deployment",
            "deploying machine learning models",
        ],
        "model_serving": [
            "model serving",
            "inference service",
            "online inference",
            "real-time inference",
        ],
        "inference_optimization": [
            "inference optimization",
            "latency optimization",
            "quantization",
            "distillation",
            "onnx",
            "tensorrt",
        ],
        "monitoring": [
            "model monitoring",
            "data drift",
            "concept drift",
            "monitoring ml systems",
        ],
        "experiment_tracking": [
            "mlflow",
            "weights & biases",
            "wandb",
            "experiment tracking",
        ],
        "ci_cd": [
            "ci/cd",
            "continuous integration",
            "continuous delivery",
            "github actions",
            "gitlab ci",
        ],
        "docker": [
            "docker",
            "containerization",
            "контейнеризация",
        ],
        "kubernetes": [
            "kubernetes",
            "k8s",
        ],
        "api_backend": [
            "fastapi",
            "flask",
            "rest api",
            "grpc",
            "microservices",
            "backend services",
        ],
    },

    "cloud_platform": {
        "aws": [
            "aws",
            "amazon web services",
            "sagemaker",
            "bedrock",
        ],
        "gcp": [
            "gcp",
            "google cloud",
            "vertex ai",
            "google kubernetes engine",
        ],
        "azure": [
            "azure",
            "azure ml",
            "microsoft azure",
        ],
    },

    "math_stats": {
        "statistics": [
            "statistics",
            "statistical analysis",
            "математическая статистика",
            "статистика",
        ],
        "probability": [
            "probability theory",
            "theory of probability",
            "теория вероятностей",
        ],
        "linear_algebra": [
            "linear algebra",
            "линейная алгебра",
        ],
    },

    "soft_and_process": {
        "communication": [
            "communication",
            "cross-functional",
            "stakeholder management",
            "коммуникация",
            "взаимодействие с командами",
        ],
        "leadership": [
            "leadership",
            "mentoring",
            "mentorship",
            "team lead",
            "technical leadership",
            "руководство",
            "менторство",
        ],
        "product_thinking": [
            "product thinking",
            "product mindset",
            "business impact",
            "product metrics",
        ],
        "research": [
            "research",
            "state of the art",
            "sota",
            "research mindset",
        ],
        "english": [
            "english",
            "upper-intermediate",
            "fluent english",
            "английский",
        ],
        "agile": [
            "agile",
            "scrum",
            "kanban",
        ],
    },
}


# ----------------------------
# 2. Нормализация текста
# ----------------------------

def normalize_text(text: str) -> str:
    text = str(text).lower()

    replacements = {
        "k8s": "kubernetes",
        "pytorch lightning": "pytorch_lightning",
        "scikit learn": "scikit-learn",
        "hugging face": "huggingface",
        "a/b": "ab",
        "ci/cd": "ci_cd",
    }

    for old, new in replacements.items():
        text = text.replace(old, new)

    text = re.sub(r"[\/|]+", " ", text)
    text = re.sub(r"[\(\)\[\]\{\},;:!?]+", " ", text)
    text = re.sub(r"\s+", " ", text)

    return text.strip()


# ----------------------------
# 3. Безопасный regex-поиск
# ----------------------------

def pattern_to_regex(pattern: str) -> str:
    pattern = pattern.strip().lower()
    pattern = re.escape(pattern)
    return rf"(?<!\w){pattern}(?!\w)"


def skill_found(text: str, patterns: list[str]) -> bool:
    for pattern in patterns:
        regex = pattern_to_regex(pattern)
        if re.search(regex, text):
            return True
    return False


# ----------------------------
# 4. Извлечение компетенций
# ----------------------------

def extract_competencies(text: str, taxonomy: dict) -> dict:
    text = normalize_text(text)
    result = {}

    for group_name, skills in taxonomy.items():
        found = []

        for canonical_skill, patterns in skills.items():
            if skill_found(text, patterns):
                found.append(canonical_skill)

        result[group_name] = sorted(found)

    return result


def flatten_competencies(extracted: dict) -> list[str]:
    all_skills = []
    for _, skills in extracted.items():
        all_skills.extend(skills)
    return sorted(set(all_skills))


def join_items(items: list[str]) -> str:
    return ", ".join(items) if items else ""


# ----------------------------
# 5. Агрегированные флаги
# ----------------------------

AGGREGATES = {
    "is_llm_role": ["llm", "rag", "fine_tuning", "embeddings", "prompt_engineering"],
    "is_nlp_role": ["nlp", "information_retrieval", "text_generation", "llm"],
    "is_cv_role": ["computer_vision", "ocr", "video_ml", "multimodal"],
    "is_mlops_role": ["mlops", "model_deployment", "model_serving", "monitoring", "experiment_tracking"],
    "is_data_heavy_role": ["spark", "etl_elt", "streaming", "data_warehousing", "feature_store"],
    "is_researchy_role": ["research", "deep_learning", "hyperparameter_tuning", "model_training"],
}


# ----------------------------
# 6. Применение к CSV
# ----------------------------

df = pd.read_csv("data/vacancies_labeled_ml_dl.csv")

# Подставь свою колонку с текстом вакансии
TEXT_COLUMNS_CANDIDATES = ["description", "vacancy_text", "text", "body"]

text_col = None
for col in TEXT_COLUMNS_CANDIDATES:
    if col in df.columns:
        text_col = col
        break

if text_col is None:
    raise ValueError(f"Не нашёл колонку с текстом. Доступные колонки: {list(df.columns)}")

# Опционально: убрать "точно нет"
if "ml_dl_label" in df.columns:
    df = df[df["ml_dl_label"] != "точно нет"].copy()

extracted_series = df[text_col].apply(lambda x: extract_competencies(x, COMPETENCY_TAXONOMY))

# Колонки по группам
for group_name in COMPETENCY_TAXONOMY.keys():
    df[group_name] = extracted_series.apply(lambda x: join_items(x[group_name]))

# Общая колонка всех компетенций
df["all_competencies"] = extracted_series.apply(lambda x: join_items(flatten_competencies(x)))

# One-hot колонки
for group_name, skills in COMPETENCY_TAXONOMY.items():
    for skill_name in skills.keys():
        col_name = f"skill__{skill_name}"
        df[col_name] = extracted_series.apply(lambda x: int(skill_name in x[group_name]))

# Агрегированные флаги
for agg_name, agg_skills in AGGREGATES.items():
    df[agg_name] = extracted_series.apply(
        lambda x: int(any(skill in flatten_competencies(x) for skill in agg_skills))
    )

df = df.drop(columns=['vacancy_text', 'ml_dl_label', 'ml_dl_binary', 'confidence', 'reason'])

# Сохранение
df.to_csv("vacancies_with_competencies.csv", index=False)

print("Готово.")
print("Использована колонка с текстом:", text_col)
print("Размер итогового датафрейма:", df.shape)
```

```python
from collections import Counter
import pandas as pd

# 1. Считаем частоты всех компетенций
counter = Counter()

for value in df["all_competencies"].fillna(""):
    skills = [x.strip() for x in value.split(",") if x.strip()]
    counter.update(skills)

# 2. Берём топ-30
top_30_skills = {skill for skill, _ in counter.most_common(30)}

# 3. В каждой строке оставляем только компетенции из топ-30
def keep_top_30_skills(value: str) -> str:
    skills = [x.strip() for x in str(value).split(",") if x.strip()]
    filtered = [skill for skill in skills if skill in top_30_skills]
    return ", ".join(filtered)

df["all_competencies"] = df["all_competencies"].apply(keep_top_30_skills)

# 4. Удаляем строки, где после фильтра ничего не осталось
df = df[df["all_competencies"].str.strip() != ""]

# 5. При желании удаляем старые one-hot колонки skill__
skill_cols = [col for col in df.columns if col.startswith("skill__")]
df = df.drop(columns=skill_cols, errors="ignore")

# 6. Создаём новые one-hot только для топ-30
for skill in sorted(top_30_skills):
    df[f"skill__{skill}"] = df["all_competencies"].apply(
        lambda x: int(skill in [s.strip() for s in str(x).split(",") if s.strip()])
    )

# 7. Сохраняем
df.to_csv("vacancies_top30_competencies.csv", index=False)
```

```python
import re
import numpy as np
import pandas as pd

# =========================
# 1. Загрузка файла
# =========================
df = pd.read_csv("vacancies_top30_competencies.csv")

# =========================
# 2. Курсы валют -> RUB
# При необходимости поправь
# =========================
CURRENCY_RATES_TO_RUB = {
    "RUB": 1.0,
    "RUR": 1.0,
    "USD": 92.0,
    "EUR": 100.0,
    "KZT": 0.20,
    "BYN": 28.0,
    "UAH": 2.2,
    "UZS": 0.0073,
    "GEL": 34.0,
    "KGS": 1.05,
    "AZN": 54.0,
}


# =========================
# 3. Определение валюты
# =========================
def detect_currency(text: str):
    if pd.isna(text):
        return None

    text = str(text).lower()

    if any(x in text for x in ["руб", "rur", "rub", "₽"]):
        return "RUB"
    if any(x in text for x in ["usd", "$", "доллар"]):
        return "USD"
    if any(x in text for x in ["eur", "€", "евро"]):
        return "EUR"
    if any(x in text for x in ["kzt", "₸", "тенге"]):
        return "KZT"
    if any(x in text for x in ["byn", "бел. руб", "белорус"]):
        return "BYN"
    if any(x in text for x in ["uah", "грн"]):
        return "UAH"
    if any(x in text for x in ["uzs", "сум"]):
        return "UZS"
    if any(x in text for x in ["gel", "лари"]):
        return "GEL"
    if any(x in text for x in ["kgs", "сом"]):
        return "KGS"
    if any(x in text for x in ["azn", "манат"]):
        return "AZN"

    return None


# =========================
# 4. Парсинг salary
# Возвращает одну оценку salary_rub
# =========================
def parse_salary_to_rub(raw_value):
    if pd.isna(raw_value):
        return np.nan

    text = str(raw_value).strip().lower()

    if text in {"", "none", "null", "nan", "не указана"}:
        return np.nan

    currency = detect_currency(text)

    # убрать пробелы внутри чисел: "150 000" -> "150000"
    text_clean = re.sub(r"(?<=\d)\s+(?=\d)", "", text)

    numbers = re.findall(r"\d+(?:[.,]\d+)?", text_clean)
    numbers = [float(x.replace(",", ".")) for x in numbers]

    if len(numbers) == 0:
        return np.nan

    if len(numbers) == 1:
        salary_value = numbers[0]
    else:
        salary_from = min(numbers[0], numbers[1])
        salary_to = max(numbers[0], numbers[1])
        salary_value = (salary_from + salary_to) / 2

    rate = CURRENCY_RATES_TO_RUB.get(currency, np.nan)
    if pd.isna(rate):
        return np.nan

    return salary_value * rate


df["salary_rub"] = df["salary"].apply(parse_salary_to_rub)


# =========================
# 5. Убираем строки без зарплаты
# =========================
df = df[df["salary_rub"].notna()].copy()


# =========================
# 6. Логарифмируем зарплату
# =========================
df["salary_rub_log"] = np.log1p(df["salary_rub"])


# =========================
# 7. Dummy по всем регионам
# =========================
df["region"] = df["region"].astype(str).str.strip().str.lower()
df["region"] = df["region"].replace({
    "nan": "unknown",
    "none": "unknown",
    "": "unknown",
})

region_dummies = pd.get_dummies(df["region"], prefix="region", dtype="int8")
df = pd.concat([df, region_dummies], axis=1)


# =========================
# 8. Чистим year_published
# =========================
df["year_published"] = pd.to_numeric(df["year_published"], errors="coerce")
df = df[df["year_published"].notna()].copy()
df["year_published"] = df["year_published"].astype("int16")


# =========================
# 9. Приводим бинарные признаки к int8
# =========================
binary_cols = [
    col for col in df.columns
    if col.startswith("skill__") or col.startswith("is_")
]

for col in binary_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0).astype("int8")


# =========================
# 10. Удаляем лишние колонки
# =========================
drop_cols = [
    "vacancy_id",
    "vacancy_url",
    "salary",
    "region",
    "programming",
    "ml_dl_core",
    "llm_nlp",
    "cv_audio_multimodal",
    "ml_domains",
    "frameworks_libraries",
    "data_engineering",
    "mlops_production",
    "cloud_platform",
    "math_stats",
    "soft_and_process",
    "all_competencies",
]

df = df.drop(columns=[c for c in drop_cols if c in df.columns], errors="ignore")


# =========================
# 11. Удаляем дубликаты
# =========================
df = df.drop_duplicates().copy()


# =========================
# 12. Сохраняем
# =========================
df.to_csv("vacancies_top30_competencies_train_ready.csv", index=False)

print("Итоговый размер:", df.shape)
print(df.columns.tolist())
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from catboost import CatBoostRegressor

# =========================
# 1. Загрузка данных
# =========================
df = pd.read_csv("vacancies_top30_competencies_train_ready.csv")

# =========================
# 2. Target и признаки
# =========================
TARGET_COL = "salary_rub_log"

if TARGET_COL not in df.columns:
    raise ValueError(f"Не найдена колонка {TARGET_COL}")

X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL].copy()

# На всякий случай удалим salary_rub, если хочешь предсказывать именно log
# Иначе это жёсткая утечка, потому что salary_rub = expm1(salary_rub_log)
if "salary_rub" in X.columns:
    X = X.drop(columns=["salary_rub"])

# =========================
# 3. train/valid/test split
# =========================
X_train_full, X_test, y_train_full, y_test = train_test_split(
    X,
    y,
    test_size=0.15,
    random_state=42
)

X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full,
    y_train_full,
    test_size=0.15,
    random_state=42
)

print("Train shape:", X_train.shape)
print("Valid shape:", X_valid.shape)
print("Test shape:", X_test.shape)

# =========================
# 4. CatBoost
# =========================
model = CatBoostRegressor(
    loss_function="RMSE",
    eval_metric="RMSE",
    iterations=2000,
    learning_rate=0.03,
    depth=6,
    l2_leaf_reg=5,
    random_seed=42,
    verbose=200,
    early_stopping_rounds=100
)

model.fit(
    X_train,
    y_train,
    eval_set=(X_valid, y_valid),
    use_best_model=True
)

# =========================
# 5. Предсказания
# =========================
valid_pred = model.predict(X_valid)
test_pred = model.predict(X_test)

# =========================
# 6. Метрики в лог-шкале
# =========================
def regression_metrics(y_true, y_pred, prefix=""):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"{prefix}RMSE: {rmse:.4f}")
    print(f"{prefix}MAE:  {mae:.4f}")
    print(f"{prefix}R2:   {r2:.4f}")

print("\nValidation metrics (log scale):")
regression_metrics(y_valid, valid_pred, prefix="valid_")

print("\nTest metrics (log scale):")
regression_metrics(y_test, test_pred, prefix="test_")

# =========================
# 7. Метрики в рублях
# =========================
y_test_rub = np.expm1(y_test)
test_pred_rub = np.expm1(test_pred)

rub_rmse = np.sqrt(mean_squared_error(y_test_rub, test_pred_rub))
rub_mae = mean_absolute_error(y_test_rub, test_pred_rub)

print("\nTest metrics (rub scale):")
print(f"test_RMSE_rub: {rub_rmse:.2f}")
print(f"test_MAE_rub:  {rub_mae:.2f}")

# =========================
# 8. Важность признаков
# =========================
feature_importance = pd.DataFrame({
    "feature": X_train.columns,
    "importance": model.get_feature_importance()
}).sort_values("importance", ascending=False)

print("\nTop 30 features:")
print(feature_importance.head(30))

# =========================
# 9. Сохранение
# =========================
model.save_model("catboost_salary_log_model.cbm")
feature_importance.to_csv("catboost_salary_log_feature_importance.csv", index=False)

# Предсказания на тесте
test_predictions = X_test.copy()
test_predictions["y_true_log"] = y_test.values
test_predictions["y_pred_log"] = test_pred
test_predictions["y_true_rub"] = y_test_rub.values
test_predictions["y_pred_rub"] = test_pred_rub

test_predictions.to_csv("catboost_salary_log_test_predictions.csv", index=False)

print("\nМодель сохранена в catboost_salary_log_model.cbm")
```

```python
import numpy as np
import pandas as pd

from catboost import CatBoostRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv("vacancies_top30_competencies_train_ready.csv")

# только Москва и СПб
df = df[
    (df["region_москва"] == 1) |
    (df["region_санкт-петербург"] == 1)
].copy()

# оставить только нужные region-колонки
keep_region_cols = ["region_москва", "region_санкт-петербург"]
all_region_cols = [c for c in df.columns if c.startswith("region_")]
drop_region_cols = [c for c in all_region_cols if c not in keep_region_cols]
df = df.drop(columns=drop_region_cols, errors="ignore")

# убрать is_*
df = df.drop(columns=[c for c in df.columns if c.startswith("is_")], errors="ignore")

# убрать выбросы
low = df["salary_rub_log"].quantile(0.01)
high = df["salary_rub_log"].quantile(0.99)
df = df[(df["salary_rub_log"] >= low) & (df["salary_rub_log"] <= high)].copy()

# убрать редкие skills
skill_cols = [c for c in df.columns if c.startswith("skill__")]
rare_skill_cols = [c for c in skill_cols if df[c].sum() < 20]
df = df.drop(columns=rare_skill_cols, errors="ignore")

y = df["salary_rub_log"].copy()

base_cols = ["year_published", "region_москва", "region_санкт-петербург"]
base_cols = [c for c in base_cols if c in df.columns]

skill_cols = [c for c in df.columns if c.startswith("skill__")]

X_base = df[base_cols].copy()
X_full = df[base_cols + skill_cols].copy()

kf = KFold(n_splits=5, shuffle=True, random_state=42)

def evaluate_model(X, y, name):
    rmse_scores, mae_scores, r2_scores = [], [], []

    for train_idx, valid_idx in kf.split(X):
        X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
        y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

        model = CatBoostRegressor(
            loss_function="RMSE",
            eval_metric="RMSE",
            iterations=2000,
            learning_rate=0.03,
            depth=6,
            l2_leaf_reg=5,
            random_seed=42,
            verbose=0,
            early_stopping_rounds=100
        )

        model.fit(
            X_train,
            y_train,
            eval_set=(X_valid, y_valid),
            use_best_model=True
        )

        pred = model.predict(X_valid)

        rmse_scores.append(np.sqrt(mean_squared_error(y_valid, pred)))
        mae_scores.append(mean_absolute_error(y_valid, pred))
        r2_scores.append(r2_score(y_valid, pred))

    print(f"\n{name}")
    print(f"RMSE mean = {np.mean(rmse_scores):.4f} ± {np.std(rmse_scores):.4f}")
    print(f"MAE  mean = {np.mean(mae_scores):.4f} ± {np.std(mae_scores):.4f}")
    print(f"R2   mean = {np.mean(r2_scores):.4f} ± {np.std(r2_scores):.4f}")

evaluate_model(X_base, y, "BASELINE: region + year")
evaluate_model(X_full, y, "FULL: region + year + skills")
```

```python
import re
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =========================================
# 1. Загрузка данных
# =========================================
df = pd.read_csv("vacancies_ml_related_with_salary.csv")

print("Исходный размер:", df.shape)
print(df.columns.tolist())


# =========================================
# 2. Курсы валют к рублю
# При необходимости скорректируй
# =========================================
CURRENCY_RATES_TO_RUB = {
    "RUB": 1.0,
    "RUR": 1.0,
    "USD": 92.0,
    "EUR": 100.0,
    "KZT": 0.20,
    "BYN": 28.0,
    "UAH": 2.2,
    "UZS": 0.0073,
    "GEL": 34.0,
    "KGS": 1.05,
    "AZN": 54.0,
}


# =========================================
# 3. Определение валюты
# =========================================
def detect_currency(text: str):
    if pd.isna(text):
        return None

    text = str(text).lower()

    if any(x in text for x in ["руб", "rur", "rub", "₽"]):
        return "RUB"
    if any(x in text for x in ["usd", "$", "доллар"]):
        return "USD"
    if any(x in text for x in ["eur", "€", "евро"]):
        return "EUR"
    if any(x in text for x in ["kzt", "₸", "тенге"]):
        return "KZT"
    if any(x in text for x in ["byn", "бел. руб", "белорус"]):
        return "BYN"
    if any(x in text for x in ["uah", "грн"]):
        return "UAH"
    if any(x in text for x in ["uzs", "сум"]):
        return "UZS"
    if any(x in text for x in ["gel", "лари"]):
        return "GEL"
    if any(x in text for x in ["kgs", "сом"]):
        return "KGS"
    if any(x in text for x in ["azn", "манат"]):
        return "AZN"

    return None


# =========================================
# 4. Парсинг зарплаты в RUB
# =========================================
def parse_salary_to_rub(raw_value):
    if pd.isna(raw_value):
        return np.nan

    text = str(raw_value).strip().lower()

    if text in {"", "none", "null", "nan", "не указана"}:
        return np.nan

    currency = detect_currency(text)

    # "150 000" -> "150000"
    text_clean = re.sub(r"(?<=\d)\s+(?=\d)", "", text)

    numbers = re.findall(r"\d+(?:[.,]\d+)?", text_clean)
    numbers = [float(x.replace(",", ".")) for x in numbers]

    if len(numbers) == 0:
        return np.nan

    if len(numbers) == 1:
        salary_value = numbers[0]
    else:
        salary_from = min(numbers[0], numbers[1])
        salary_to = max(numbers[0], numbers[1])
        salary_value = (salary_from + salary_to) / 2

    rate = CURRENCY_RATES_TO_RUB.get(currency, np.nan)
    if pd.isna(rate):
        return np.nan

    return salary_value * rate


df["salary_rub"] = df["salary"].apply(parse_salary_to_rub)
df["salary_rub_log"] = np.log1p(df["salary_rub"])


# =========================================
# 5. Фильтрация данных
# Оставляем только ML/DL и строки с текстом и зарплатой
# =========================================
df = df[df["ml_dl_label"] != "точно нет"].copy()
df = df[df["vacancy_text"].notna()].copy()
df = df[df["salary_rub"].notna()].copy()
df = df[df["salary_rub"] > 0].copy()

# можно убрать экстремальные выбросы
low = df["salary_rub_log"].quantile(0.01)
high = df["salary_rub_log"].quantile(0.99)
df = df[(df["salary_rub_log"] >= low) & (df["salary_rub_log"] <= high)].copy()

print("После фильтрации:", df.shape)


# =========================================
# 6. Удаление ненужных колонок
# Для обучения по тексту нам нужен текст и target
# =========================================
drop_cols = [
    "vacancy_id",
    "vacancy_url",
    "salary",
    "region",
    "year_published",
    "ml_dl_label",
    "ml_dl_binary",
    "confidence",
    "reason",
]

df = df.drop(columns=[c for c in drop_cols if c in df.columns], errors="ignore")

print("Колонки после чистки:", df.columns.tolist())


# =========================================
# 7. Признаки и target
# =========================================
X = df["vacancy_text"].astype(str)
y = df["salary_rub_log"].astype(float)


# =========================================
# 8. Модель: TF-IDF + Ridge
# =========================================
kf = KFold(n_splits=5, shuffle=True, random_state=42)

rmse_scores = []
mae_scores = []
r2_scores = []

for fold, (train_idx, valid_idx) in enumerate(kf.split(X), 1):
    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

    model = Pipeline([
        ("tfidf", TfidfVectorizer(
            lowercase=True,
            max_features=30000,
            ngram_range=(1, 2),
            min_df=3,
            max_df=0.95
        )),
        ("regressor", Ridge(alpha=3.0))
    ])

    model.fit(X_train, y_train)
    pred = model.predict(X_valid)

    rmse = np.sqrt(mean_squared_error(y_valid, pred))
    mae = mean_absolute_error(y_valid, pred)
    r2 = r2_score(y_valid, pred)

    rmse_scores.append(rmse)
    mae_scores.append(mae)
    r2_scores.append(r2)

    print(f"Fold {fold}: RMSE={rmse:.4f}, MAE={mae:.4f}, R2={r2:.4f}")

print("\nCV results:")
print(f"RMSE mean = {np.mean(rmse_scores):.4f} ± {np.std(rmse_scores):.4f}")
print(f"MAE  mean = {np.mean(mae_scores):.4f} ± {np.std(mae_scores):.4f}")
print(f"R2   mean = {np.mean(r2_scores):.4f} ± {np.std(r2_scores):.4f}")


# =========================================
# 9. Обучение финальной модели на всех данных
# =========================================
final_model = Pipeline([
    ("tfidf", TfidfVectorizer(
        lowercase=True,
        max_features=30000,
        ngram_range=(1, 2),
        min_df=3,
        max_df=0.95
    )),
    ("regressor", Ridge(alpha=3.0))
])

final_model.fit(X, y)

print("\nФинальная модель обучена.")
```

```python
import re
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =========================================
# 1. Загрузка данных
# =========================================
df = pd.read_csv("vacancies_ml_related_with_salary.csv")

print("Исходный размер:", df.shape)


# =========================================
# 2. Курсы валют к рублю
# =========================================
CURRENCY_RATES_TO_RUB = {
    "RUB": 1.0,
    "RUR": 1.0,
    "USD": 92.0,
    "EUR": 100.0,
    "KZT": 0.20,
    "BYN": 28.0,
    "UAH": 2.2,
    "UZS": 0.0073,
    "GEL": 34.0,
    "KGS": 1.05,
    "AZN": 54.0,
}


# =========================================
# 3. Определение валюты
# =========================================
def detect_currency(text: str):
    if pd.isna(text):
        return None

    text = str(text).lower()

    if any(x in text for x in ["руб", "rur", "rub", "₽"]):
        return "RUB"
    if any(x in text for x in ["usd", "$", "доллар"]):
        return "USD"
    if any(x in text for x in ["eur", "€", "евро"]):
        return "EUR"
    if any(x in text for x in ["kzt", "₸", "тенге"]):
        return "KZT"
    if any(x in text for x in ["byn", "бел. руб", "белорус"]):
        return "BYN"
    if any(x in text for x in ["uah", "грн"]):
        return "UAH"
    if any(x in text for x in ["uzs", "сум"]):
        return "UZS"
    if any(x in text for x in ["gel", "лари"]):
        return "GEL"
    if any(x in text for x in ["kgs", "сом"]):
        return "KGS"
    if any(x in text for x in ["azn", "манат"]):
        return "AZN"

    return None


# =========================================
# 4. Парсинг зарплаты в RUB
# =========================================
def parse_salary_to_rub(raw_value):
    if pd.isna(raw_value):
        return np.nan

    text = str(raw_value).strip().lower()

    if text in {"", "none", "null", "nan", "не указана"}:
        return np.nan

    currency = detect_currency(text)

    text_clean = re.sub(r"(?<=\d)\s+(?=\d)", "", text)
    numbers = re.findall(r"\d+(?:[.,]\d+)?", text_clean)
    numbers = [float(x.replace(",", ".")) for x in numbers]

    if len(numbers) == 0:
        return np.nan

    if len(numbers) == 1:
        salary_value = numbers[0]
    else:
        salary_from = min(numbers[0], numbers[1])
        salary_to = max(numbers[0], numbers[1])
        salary_value = (salary_from + salary_to) / 2

    rate = CURRENCY_RATES_TO_RUB.get(currency, np.nan)
    if pd.isna(rate):
        return np.nan

    return salary_value * rate


df["salary_rub"] = df["salary"].apply(parse_salary_to_rub)
df["salary_rub_log"] = np.log1p(df["salary_rub"])


# =========================================
# 5. Фильтрация
# =========================================
df = df[df["ml_dl_label"] != "точно нет"].copy()
df = df[df["vacancy_text"].notna()].copy()
df = df[df["salary_rub"].notna()].copy()
df = df[df["salary_rub"] > 0].copy()

low = df["salary_rub_log"].quantile(0.01)
high = df["salary_rub_log"].quantile(0.99)
df = df[(df["salary_rub_log"] >= low) & (df["salary_rub_log"] <= high)].copy()

print("После фильтрации:", df.shape)


# =========================================
# 6. Оставляем только нужные колонки
# =========================================
keep_cols = [
    "vacancy_text",
    "region",
    "salary_rub_log",
]

df = df[keep_cols].copy()

# нормализация табличных колонок
df["region"] = df["region"].astype(str).str.strip().str.lower()

# можно оставить только Москва/СПб, если хочешь
df = df[df["region"].isin(["москва", "санкт-петербург"])].copy()

print("Колонки для модели:", df.columns.tolist())


# =========================================
# 7. X и y
# =========================================
X = df[["vacancy_text", "region"]].copy()
y = df["salary_rub_log"].astype(float).copy()


# =========================================
# 8. Препроцессинг
# =========================================
text_features = "vacancy_text"
categorical_features = ["region"]

preprocessor = ColumnTransformer(
    transformers=[
        (
            "text",
            TfidfVectorizer(
                lowercase=True,
                max_features=30000,
                ngram_range=(1, 2),
                min_df=3,
                max_df=0.95
            ),
            text_features
        ),
        (
            "cat",
            Pipeline([
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("onehot", OneHotEncoder(handle_unknown="ignore"))
            ]),
            categorical_features
        ),
    ],
    remainder="drop"
)


# =========================================
# 9. Модель
# =========================================
model = Pipeline([
    ("preprocessor", preprocessor),
    ("regressor", Ridge(alpha=3.0))
])


# =========================================
# 10. KFold CV
# =========================================
kf = KFold(n_splits=5, shuffle=True, random_state=42)

rmse_scores = []
mae_scores = []
r2_scores = []

for fold, (train_idx, valid_idx) in enumerate(kf.split(X), 1):
    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

    model.fit(X_train, y_train)
    pred = model.predict(X_valid)

    rmse = np.sqrt(mean_squared_error(y_valid, pred))
    mae = mean_absolute_error(y_valid, pred)
    r2 = r2_score(y_valid, pred)

    rmse_scores.append(rmse)
    mae_scores.append(mae)
    r2_scores.append(r2)

    print(f"Fold {fold}: RMSE={rmse:.4f}, MAE={mae:.4f}, R2={r2:.4f}")

print("\nCV results:")
print(f"RMSE mean = {np.mean(rmse_scores):.4f} ± {np.std(rmse_scores):.4f}")
print(f"MAE  mean = {np.mean(mae_scores):.4f} ± {np.std(mae_scores):.4f}")
print(f"R2   mean = {np.mean(r2_scores):.4f} ± {np.std(r2_scores):.4f}")


# =========================================
# 11. Финальная модель на всех данных
# =========================================
model.fit(X, y)
print("\nФинальная модель обучена.")
```
